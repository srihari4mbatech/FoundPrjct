```python

# Specify the path to your ChromeDriver executable
chrome_driver_path = "./chromedriver-win64/chromedriver.exe"

# Specify the Google News link
google_news_link = f_link
# Set up Chrome options to run in headless mode
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')

```

```python
csv_file_path = "./news_links.csv"
with open(csv_file_path, "w", encoding='utf-8-sig') as document:
    document.write("Title,Time,Description,Source\n")
```

```python
with webdriver.Chrome(
    service=ChromeService(executable_path=chrome_driver_path),
    options=chrome_options
) as driver:
    # Navigate to the Google News link
    driver.get(google_news_link)

    # Wait for a few seconds to allow the page to load
    time.sleep(5)
    while True:
        # Get the page source after JavaScript rendering
        if driver.window_handles:  # Check if there are open windows
            page_source = driver.page_source

            # Parse the HTML content using BeautifulSoup
            soup = BeautifulSoup(page_source, 'html.parser')

            # Find and print all div elements with class 'SoaBEf'
            for item in soup.find_all('div', class_='SoaBEf'):
                news_link = item.find('a', href=True)['href']
                title = item.find('div', class_='n0jPhd ynAwRc MBeuO nDgy9d').get_text(strip=True).replace(",", "").replace("\n", " ")
                time_text = item.find('div', class_='OSrXXb rbYSKb LfVVr').get_text(strip=True)
                description = item.find('div', class_='GI74Re nDgy9d').get_text(strip=True).replace(",", "").replace("\n", " ")

                # Use 'utf-8-sig' encoding to handle CSV encoding issues
                with open(csv_file_path, "a", encoding='utf-8-sig') as document:
                    document.write("{},{},{},{}\n".format(title, time_text, description, news_link))

            # Find and click on the next page link
            try:
                next_link = driver.find_element("link text", "Next")
                next_link.click()
            except NoSuchElementException:
                break
```
